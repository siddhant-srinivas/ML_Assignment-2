```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sn
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
X_train = X_train/255
X_test = X_test/255
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(150, activation='relu'),
    keras.layers.Dense(150, activation='relu'),
    keras.layers.Dense(10 ,activation='softmax')
])
model.compile(
    optimizer='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy']
)

m = model.fit(X_train, y_train, epochs=10, validation_data=(X_test,y_test))
print("Accuracy=", m.history['val_accuracy'][9])
predictions = model.predict(X_test)
predictions_labels = [np.argmax(i) for i in predictions]
cm = tf.math.confusion_matrix(labels=y_test,predictions=predictions_labels)

plt.figure(figsize= (10,7))
sn.heatmap(cm, annot=True, fmt='d', cmap='copper')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
